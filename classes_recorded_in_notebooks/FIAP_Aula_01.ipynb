# Iniciando a Receita

```python
print("PRONTOS PARA COZINHAR?")
```

## Passo 1: Ingredientes (Variáveis)

```python
print("Passo 1: Ingredientes | Adcione os ingredientes (Variáveis)\n")

# Representação de variáveis como farinha, ovos, açúcar, etc.
farinha = "farinha"
ovos = 3
acucar = "açúcar"
leite = "leite"

# Mensagem indicando a conclusão do passo
print(f"Os seguintes ingredientes foram adicionados com sucesso:\n"
      f"\n 1-Farinha: {farinha}\n 2-Ovos: {ovos}\n 3-Açucar: {acucar}\n 4-Leite: {leite}")
```

## Passo 2: Preparação (Código)

```python
print("Passo 2: Preparação | Adicione manteiga e fermento (Código)\n")

# Criação de variáveis para representar os ingredientes
manteiga = "manteiga"
fermento = "fermento"

# Demonstração de operações básicas utilizando operadores em Python
mistura = farinha + " + " + manteiga + " + " + acucar

# Mensagem indicando a conclusão do passo
print(f"A mistura contém: {mistura}")
print("Realizada com sucesso!")
```

## Passo 3: Instruções (Lógica de Programação)

```python
print("\nPasso 3: Instruções (Lógica de Programação)\n")

# Ênfase na importância da sequência de instruções na programação
print(f"Adicionamos {ovos} ovos")
if ovos >= 3:
    print("Escolha adequada de ovos.")
else:
    print("Adicione mais ovos à receita.")

# Mensagem indicando a conclusão do passo
print("Lógica de programação aplicada com sucesso!")
```

## Passo 4: Cozimento (Execução do Código)

```python
print("\nPasso 4: Cozimento (Execução do Código)\n")

# Execução do código para "cozinhar" o bolo
tempo_de_cozimento = 30  # em minutos
print(f"Tempo de cozimento é de {tempo_de_cozimento} minutos.")

# Exemplificação da depuração (ajustar o tempo de cozimento)
if tempo_de_cozimento > 25:
    print("O bolo está pronto para ser degustado!")
else:
    print("Ajuste o tempo de cozimento e execute novamente.")

# Mensagem indicando a conclusão do passo
print("A receita foi executada com sucesso!")

```

## Passo 5: Degustação (Avaliação do Resultado)

```python
print("\nPasso 5: Degustação (Avaliação do Resultado)\n")

# Análise dos outputs gerados
resultado_final = "Bolo delicioso!"

# Verificação se o resultado corresponde ao esperado
if resultado_final == "Bolo delicioso!":
    print("A apresentação está fantástica!")
    print("O sabor está magnífico!")
    print("Parabéns! Você é um Master Chef! #sqn")
else:
    print("Ajuste a receita para obter um resultado mais satisfatório.")

# Mensagem indicando a conclusão do passo
print("\n  -- FIM --  ")
```
